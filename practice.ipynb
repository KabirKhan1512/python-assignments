## Assignment 1: 

Advanced Calculator with Memory and Error Handling 

- Implement a calculator using functions that performs basic arithmetic operations. 

- Include memory functions (`M+`, `M-`, `MR`, `MC`) to store and recall results. 

- Use conditional statements to handle division by zero and other possible errors. 

- Incorporate `args` and `kwargs` to handle multiple operations in a single function.  Kabir Ud Din

```python
def add(*args):
    sums = 0
    for i in args:
        sums += i
    return sums

def subtract(*args):
    if not args:
        return 0
    result = args[0]
    for i in args[1:]:
        result -= i
    return result

def divide(*args):
    if not args:
        return 0
    result = args[0]
    for i in args[1:]:
        if i == 0:
            print("Infinity")
            break
        else:
            result /= i
    return result

def multiply(*args):
    if not args:
        return 0
    result = args[0]
    for i in args[1:]:
        result *= i
    return result

memory = 0

def m_plus(amount):
    global memory
    memory += amount
    print(memory)

def m_minus(amount):
    global memory
    memory -= amount
    print(memory)

def m_recal():
    global memory
    return memory
    print(memory)
    
def m_clear():
    global memory
    memory = 0
    print(memory)
    return memory
    
```

## Assignment 5: 
Text File Analyzer 

- Create a script that reads a text file and counts the occurrences of each word. 

- Implement functions to calculate the total number of words, the most frequent word, and the average word length. 

- Use dictionaries to store word frequencies and loops to iterate through the text. 

- Include a function to find and replace a word in the text.



```python
def occurence_counter(file_path):
    with open(file_path, "r") as file:
        text = file.read()
        text_list = text.split()
        
    counter = {}
    for i in text_list:
        if i not in counter:
            counter[i] = 1
        else:
            counter[i] += 1
    return counter

def word_replacer(file_path, word, replacement):
    with open(file_path, "r") as file:
        text = file.read()
        text.replace(word, replacement)
    print(f"The word '{word}' is replaced with '{replacement}'.")

def word_counter(file_path):
    with open(file_path, "r") as file:
        text = file.read()
        text_list = text.split()
    return len(text_list)

def most_repeated(dict):
    max_key = max(dict, key=dict.get)
    return max_key

def avg_of_words(file_path, total_words):
    with open(file_path, "r") as file:
            text = file.read()
            text_list = text.split()
    sum_of_len = 0
    for i in text_list:
        sum_of_len += len(i)
    avg = sum_of_len / total_words
    return avg
```

## Assignment 6: 
Number Guessing Game 

- Implement a number guessing game where the player has to guess a randomly generated number between 1 and 100. 

- Use loops to allow multiple attempts and provide hints ("higher" or "lower"). 

- Include a scoring system based on the number of attempts. 

- Use conditional statements to check the validity of the player's input. 

```python
import random

number_to_guess = random.randint(1, 100)


print("Welcome to Number Guessing Game! Here you have to guess the number in 5 attempts to win.")

for i in range(1, 7):
    if i == 6:
        print("You lost the game!")
        break
    number = int(input(f"Give your try no. {i}: "))
    if number == number_to_guess:
        print(f"You won your game on try no. {i}")
        break
    elif number > number_to_guess:
        print("Guess a lower number")
    elif number < number_to_guess:
        print("Guess a higher number")
```

## Assignment 7: 
Personal Budget Manager 

- Store budget categories and their respective amounts in a dictionary. 

- Implement functions to add, remove, update, and list budget categories. 

- Include a function to calculate the total budget and the remaining balance after expenses. 

- Use loops and conditional statements to manage the budget. 

```python
def create_budget_record(name):
    name = {}
    return name

def add_budget(dict_name, key_budget, value_budget):
    dict_name.update({key_budget: value_budget})

def remove_budget(dict_name, key_budget):
    del dict_name[key_budget]

def print_budget(dict_name):
    for i,j in dict_name.items():
        print(f"{i} = {j}")

def total_budget(dict_name):
    tot_budget = 0
    for i,j in dict_name.items():
        tot_budget += j
    return tot_budget

def update_budget(dict_name, key_budget, value_budget):
    if key_budget in dict_name:
        dict_name[key_budget] = value_budget
        print(f"Updated '{key_budget}' with a new budget of {value_budget}")
    else:
        print(f"'{key_budget}' not found in the budget.")

def remaining_balance(dict_name, expenses):
    total = total_budget(dict_name)
    balance = total - expenses
    return balance

while True:
    "Welcome to Personal Budget Manager"
    print("""
    Select one of the following functions:
    1: Create Budget (Necessary for first time)
    2: Add Budget Category
    3: Remove Budget Category
    4: Update Budget Category
    5: Print Total Budget
    6: Check Balance
    7: Print Budget Details
    8: Exit
    """)

    choice = int(input("Input one of the following options (1-8)"))
    if choice == 1:
        name = input("Write name of your budget: ")
        dict = create_budget_record(name)
        print(f"A budget is created with name {name}")

    elif choice == 2:
        key = input("Write name of your budget category: ")
        value = int(input("Write your value: "))
        add_budget(dict_name = dict, key_budget = key, value_budget = value)

    elif choice == 3:
        key_remove = input("Which category you want to remove: ")
        remove_budget(dict_name = dict, key_budget = key_remove)


    elif choice == 4:
        key = input("Write name of your budget category: ")
        value = int(input("Write your value: "))
        add_budget(dict_name = dict, key_budget = key, value_budget = value)


    elif choice == 5:
        print(total_budget(dict))

    elif choice == 6:
        exp = int(input("Input expense occurred"))
        remaining_balance(dict_name = dict, expenses=exp)

    elif choice == 7:
        print_budget(dict_name=dict)

    elif choice == 8:
        break
```
